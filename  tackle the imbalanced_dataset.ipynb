**Imbalanced Dataset:**

A dataset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# first 5 rows of the dataframe ,
credit_card_data.head()
```

```python
# last 5 rows of the dataframe
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent


 Number of Fraudulent Transactions --> 104

```python
legit_sample = legit.sample(n=104)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) # axis =0 means concat one above other
```
